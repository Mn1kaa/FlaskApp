# FLASK 


Flask es un framework de Python usado para desarrollar aplicaciones pequeñas... en primera instancia no posee varias características que podría tener un  framework mas completo... pero conforme se vayan utilizando se le pueden ir agregando funcionalidades

## Características 

1. Usa Jinga 2 (Template Engine)
2. Es lightweight
3. No tiene ORM de forma directa...


## ¿Cómo funciona el request de una aplicación web?

Básicamente la computadora manda una petición al servidor este la lee y consulta en la base de datos por la información que se solicite... luego se trabaja esta información y se le entrega al servidor la cual devuelve con un response al navegador del usuario el cual renderiza la información HTML enviada por el servidor

## TEMPLATES

### ¿Para que sirven?

Usando Jinja2 podemos generar HTML estatico o dinamico lo cual nos permitirá bastante libertad y nos permitira regresar HTML en vez de un string...

Flask esta configurado para buscar en un directorio llamado  **templates ***


### Estructuras de Control

SINTAXIS IF :

    {% if variable %}
        <h1> Ejemplo </h1>
    {% else %}
        <h2> Ejemplo 2</>
    
    {% endif %}

SINTAXIS EXTENDER:
    {%extends 'nombre_archivo.html' %} 
* Sirve para tener un HTML base al cual modificaremos...


SINTAXIS "BLOQUE" :

    {% block title %}

    {{super()}}
    Bienvenido

    {% end block %}

    {% block content %}

    
    {% end block %}
